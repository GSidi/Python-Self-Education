```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
#df is data frame but with 2 letters ;P 
df = pd.read_csv("canada_per_capita_income.csv")
df
```

```python
%matplotlib
plt.xlabel("year")
plt.ylabel("per capita income (US$)")
plt.scatter(df.year,df.per_capita_income,color = "purple", marker = "*")
```

```python
new_df = df.drop("per_capita_income", axis="columns")
new_df
```

```python
per_capita_income = df.per_capita_income
per_capita_income
```

```python
reg = linear_model.LinearRegression()
reg.fit(new_df,per_capita_income)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
reg.predict([[2020]])
```

```python
%matplotlib inline
plt.xlabel("years", fontsize = 20)
plt.ylabel("capita", fontsize = 20)
plt.scatter(df.year,df.per_capita_income,color = "purple", marker = "+")
plt.plot(df.year, reg.predict(df[["year"]]), color = "blue")
```
